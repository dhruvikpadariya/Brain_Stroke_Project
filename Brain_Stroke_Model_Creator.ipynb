```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import pickle
```

```python
dt = pd.read_csv("stroke_risk_dataset.csv")
dt
```

```python
X = dt.iloc[:,0:17]
y = dt.iloc[:,17:]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=.3)
```

```python
sc = StandardScaler()
```

```python
X_train_std = sc.fit_transform(X_train)
```

```python
X_test_std = sc.fit_transform(X_test)
```

```python
svc_model = SVC(kernel="rbf")
```

```python
svc_model.fit(X_train_std,y_train)
```

```python
f1 = open("svc_rbf_model.pkl","wb") # wb = read write in binary format
f2 = open("std_scaler.pkl","wb") # wb = read write in binary format 
```

```python
pickle.dump(svc_model,f1,protocol=pickle.HIGHEST_PROTOCOL)
pickle.dump(sc,f2,protocol=pickle.HIGHEST_PROTOCOL)
```
